```python
from random import randint

cell_population = [10, 20, 30]
rec = [i for i in range(10)]

recordings_params = mr.get_default_recordings_params()
recordings_params['spiketrains']['duration'] = 60
recordings_params['recordings']['noise_level'] = 5

for cell in cell_population:
    print('start recording generation --------------------------------------------------', '# of cells: '+str(cell))
    k=0
    recordings_params['spiketrains']['n_exc'] = int(.7*cell)
    recordings_params['spiketrains']['n_inh'] = int(.3*cell)
    for r in rec:
        print('start --------------------------------------------------', 'setup'+str(k))
        k += 1
        recordings_params['seeds']['spiketrains'] = random.randint(0,100)
        recordings_params['seeds']['templates'] = random.randint(0,100)
        recordings_params['seeds']['noise'] = random.randint(0,100)
        recordings_params['seeds']['convolution'] = random.randint(0,100)
        recgen60 = mr.gen_recordings(params=recordings_params, templates='templates_100_SqMEA-10-15.h5')
        mr.save_recording_generator(recgen60, 'data/MEA10*10/recgen_cell_%s/%s.h5'%(cell,r))
```
